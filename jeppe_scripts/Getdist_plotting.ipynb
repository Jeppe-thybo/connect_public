```python
import matplotlib
matplotlib.use('Agg')  # Use a non-interactive backend for plotting

from getdist import plots
from getdist.mcsamples import loadMCSamples
import numpy as np

import getdist
from getdist import plots

# Load your chains
samples = getdist.loadMCSamples('/home/jeppethybo/connect_public/Grendel_files/SIDR_lu_blend_T_3/plots_T_3/2026-02-17_200000_', settings={'ignore_rows': 0.3})

# Apply temperature scaling to weights
T = 3.0
samples.weights = samples.weights ** (1.0/T)

# Renormalize weights
samples.weights = samples.weights / samples.weights.sum() * len(samples.weights)

# Now plot as usual
g = plots.get_subplot_plotter()
g.triangle_plot(samples)
g.export('/home/jeppethybo/connect_public/Grendel_files/SIDR_lu_blend_T_3/plots_T_3/triangle_plot.png')
```
